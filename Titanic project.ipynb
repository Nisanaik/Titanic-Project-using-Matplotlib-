## What is Matplotlib?

*   Matplotlib is a popular Python library for creating visualizations and plots.
 
*   Matplotlib was first released in 2003 by John D. Hunter
*   It provides a comprehensive set of tools for creating 2D plots, bar charts, histograms, scatter plots, line plots, contour plots, and many other types of visualizations.
* The library is built on top of NumPy and provides a high-level interface for creating and customizing plots.




### Plots in Matplotlib



1.   Line Plot: A line plot is a basic plot that shows the relationship between two variables. It can be created using the plot function in matplotlib.

2.   Bar Plot: A bar plot is used to compare the data between different groups or categories. It can be created using the bar function in matplotlib.

3.   Scatter Plot: A scatter plot is used to visualize the relationship between two variables. It can be created using the scatter function in matplotlib.

4.   Histogram: A histogram is used to visualize the distribution of a single variable. It can be created using the hist function in matplotlib.

5.   Pie Chart: A pie chart is used to show the proportion of different categories in a dataset. It can be created using the pie function in matplotlib.

6.   An area plot, also known as a stacked area plot, is a type of plot that displays quantitative data as a series of filled-in areas, with each area representing a different category or subgroup of the data. In Matplotlib, an area plot can be created using the fill_between() or fill_betweenx() functions.

Solution

# Line plot

```python
survived_by_year = titanic.groupby('age')['survived'].sum()
```

```python
plt.plot(survived_by_year.index,survived_by_year.values)
plt.xlabel('Year')
plt.ylabel('Number of survived')
plt.title('Line Plot for Titanic Survived by Year')
plt.show()
```

# Bar Plot

```python
survived_counts = titanic["survived"].value_counts()
```

```python
plt.bar(survived_counts.index,survived_counts.values)
plt.title("Bar Plot for Passenger Count by survived ")
plt.xlabel("survived")
plt.ylabel('count')
plt.show()
```

# Scatter Plot

```python
plt.scatter(titanic['age'], titanic['fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot for Titanic Dataset')
plt.show()
```

# Pie Plot

```python
plt.figure(figsize=(7,7))
titanic.survived.value_counts().plot(kind = 'pie',autopct='%2.1f%%')
plt.title("Pie Plot")
```

# Area Plot

```python
grouped = titanic.groupby(['pclass','survived'])['sex'].count().unstack()
grouped.plot(kind='area',stacked=True)
plt.title("passenger Survived by class")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passemgers")
plt.show()
```

# Histogram

```python
plt.hist(titanic['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Titanic Passenger Ages')

# Show the plot
plt.show()
```
